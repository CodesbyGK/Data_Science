```python
import pandas as pd

player = pd.read_csv("/content/Player.csv")
playerat = pd.read_csv("/content/Player_Attributes.csv")
```

```python
playerat.shape
```

```python
merged_df = pd.merge(player, playerat, on=['player_api_id', 'player_fifa_api_id'], how='inner')
display(merged_df.head())
```

```python
merged_df.drop_duplicates(subset=['player_name'], inplace=True)
display(merged_df.head())
merged_df.shape
```

```python
merged_df = merged_df.drop('id_y', axis=1)
display(merged_df)
```
